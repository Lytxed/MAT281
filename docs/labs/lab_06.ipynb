El **cuarteto de Anscombe** comprende cuatro conjuntos de datos que tienen las mismas propiedades estadísticas, pero que evidentemente son distintas al inspeccionar sus gráficos respectivos.

Cada conjunto consiste de once puntos (x, y) y fueron construidos por el estadístico F. J. Anscombe. El cuarteto es una demostración de la importancia de mirar gráficamente un conjunto de datos antes de analizarlos.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split



%matplotlib inline
sns.set_palette("deep", desat=.6)
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
# Cargar los datos del cuarteto de Anscombe
data = sns.load_dataset("anscombe")
data.head()
```

Basado en la información presentada responda las siguientes preguntas:

1. Gráfique mediante un gráfico tipo **scatter** cada grupo. A simple vista, ¿ los grupos son muy distintos entre si?.

```python
df_1=data[data["dataset"]=='I']
df_2=data[data["dataset"]=='II']
df_3=data[data["dataset"]=='III']
df_4=data[data["dataset"]=='IV']

#Graficos de scatter para cada df
plt.figure(figsize=(8, 4))
sns.scatterplot(
    x='x',
    y='y',
    hue='dataset', #Se crea una leyenda para cada dataset
    data=df_1, #Se ocupa la informa de df_1
)
plt.title('Gráfico tipo scatter de df_1')
plt.show()
```

3. Realice un ajuste lineal (con **sklearn**) y calcule los resultados de las métricas para cada grupo. Además, grafique los resultados de la regresión lineal para cada grupo. Interprete.

```python

X = np.array(df_1[['x']])
y = np.array(df_1['y'])

# Crea un modelo de regresión lineal
modelo = LinearRegression()

# Ajusta el modelo a los datos
modelo.fit(X, y)

# Realiza predicciones
predicciones = modelo.predict(X)

# Imprime los coeficientes y el intercepto
print("Coeficiente:", modelo.coef_)
print("Intercepto:", modelo.intercept_)

# Graficar los datos originales y la línea de regresión
plt.figure(figsize=(8, 4))
plt.scatter(X, y, color='blue', label='Datos')
plt.plot(X, predicciones, color='red', label='Línea de Regresión')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
r2_score(np.array(y), predicciones)
```

4. Es claro que el ajuste lineal para algunos grupos no es el correcto. Existen varias formas de solucionar este problema (eliminar outliers, otros modelos, etc.). Identifique una estrategia para que el modelo de regresión lineal ajuste de mejor manera e implemente otros modelos en los casos que encuentre necesario.  

Notemos que para los datos de df_1 son muy aleatorios y el ajuste lineal es el mejor para este caso. Por otro lado, para los datos df_4 no se puede mejorar ya que hay muchos datos en una linea vertical y no otro ajuste que sea mejor.

```python
x=df_2['x']
y=df_2['y']
line = np.linspace(3,15,30)
polymodel = np.poly1d(np.polyfit(x, y, 2)) #hacemos un ajuste de polinomio de grado 2

plt.scatter(x, y) #graficamos los puntos
plt.plot(line, polymodel(line)) #graficamos el polinomio
plt.show()

r2_score(np.array(y), np.array(polymodel(x))) #además, podemos ver el r^2 para el reajuste
```

```python
X = np.array(df_3.drop(24)[['x']]) #quiamos el dato
y = np.array(df_3.drop(24)['y'])

# Crea un modelo de regresión lineal
modelo = LinearRegression()

# Ajusta el modelo a los datos
modelo.fit(X, y)

# Realiza predicciones
predicciones = modelo.predict(X)

# Imprime los coeficientes y el intercepto
print("Coeficiente:", modelo.coef_)
print("Intercepto:", modelo.intercept_)

# Graficar los datos originales y la línea de regresión
plt.figure(figsize=(8, 4))
plt.scatter(X, y, color='blue', label='Datos')
plt.plot(X, predicciones, color='red', label='Línea de Regresión')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
r2_score(np.array(y), predicciones) #Coeficiente que dice que si el ajuste es mejor
```
