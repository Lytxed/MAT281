3. Desarrolle una función `resumen_df(df)` para encontrar el total de elementos distintos y vacíos por columnas.

```python
# respuesta

def resumen_df(df):

    result = pd.DataFrame({'nombres': df.columns})
    result['elementos_distintos']=list(df.nunique())
    result['elementos_vacios'] = list(df.isnull().sum())

    return result
```

```python
resumen_df(df)
```

4. Para los paises latinoamericano, encuentre por año  el país con mayor y menor `indice`.

 * a) Mediante un ciclo _for_.
 * b) Mediante un  _groupby_.

```python
# respuesta
america = ['ARG', 'ATG', 'BLZ', 'BOL', 'BRA', 'CAN', 'CHL', 'COL', 'CRI',
       'CUB', 'DOM', 'ECU', 'GRD', 'GTM', 'GUY', 'HND', 'HTI', 'JAM',
       'MEX', 'NIC', 'PAN', 'PER', 'PRY', 'SLV', 'SUR', 'TTO', 'URY',
       'USA', 'VEN']

df_america = df[df['codigo_iso'].isin(america)]
```

```python
dictAnio = dict()

for index, row in df_america.iterrows():
    if row['anio'] not in dictAnio:
        dictAnio[row['anio']] = []
    dictAnio[row['anio']] += [(row['codigo_iso'], row['pais'], row['indice'])]
```

```python
dictMaxAnio = dict()
for anio in dictAnio:
    maxIso = ''
    maxPais = ''
    maxIndice = -float('inf')

    for iso, pais, indice in dictAnio[anio]:


        if indice > maxIndice:
            maxIso = iso
            maxPais = pais
            maxIndice = indice

    dictMaxAnio[anio] = [(maxIso, maxPais, maxIndice)]
dictMaxAnio
```

Encontrar el maximo por anio mediante for

```python
dictMinAnio = dict()
for anio in dictAnio:
    minIso = ''
    minPais = ''
    minIndice = float('inf')

    for iso, pais, indice in dictAnio[anio]:


        if indice < minIndice:
            minIso = iso
            minPais = pais
            minIndice = indice

    dictMinAnio[anio] = [(minIso, minPais, minIndice)]
(dictMinAnio)
```

Encontrar el minimo por anio mediante for

```python
df_america[df_america['indice'] == df_america.groupby(['anio'])['indice'].transform('max')].sort_values('anio').reset_index()
```

```python
df_america[df_america['indice'] == df_america.groupby(['anio'])['indice'].transform('min')].sort_values('anio').reset_index()
```
