```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv("https://raw.githubusercontent.com/fralfaro/MAT281_2022/main/labs/lab_04/data/company_sales_data.csv")
df.head()
```

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
lista=[]
x= np.array(range(1,len(df['total_profit'])+1))
for i in range(0, len(df['total_profit'])):
  lista.append(df['total_profit'][i])

# grafico
fig = plt.figure(figsize=(10, 5)) # ventana
plt.scatter(
    x, # eje x
    lista, # eje y
    c="black", # color
    alpha=0.7, # intensidad color
    label="Ganancias" # etiquetdo fijando posicion
)
plt.xlabel("Meses")
plt.ylabel("Ganancias Mensual")
plt.legend(loc='upper left')
plt.show()
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
lista=[]
x= np.array(range(1,len(df['facecream'])+1))
for i in range(0, len(df['facecream'])):
  lista.append(df['facecream'][i])

fig = plt.figure(figsize=(10, 5))
plt.bar(
    x, # eje x
    lista, # eje y
    align='center', # centrar nombre eje x
    color="blue", # color
    alpha=0.6 # intensidad del color
)

plt.xticks(x, x)
plt.title('Venta de Face Cream')
plt.xlabel('Meses')
plt.ylabel('Venta mensual')
plt.show()
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
stats_df = df.drop(['month_number', 'total_units', 'total_profit'], axis=1) #bota las columnas que no quiero utilizar

plt.figure(figsize=(10, 6))
sns.boxplot(data=stats_df)
plt.title('Ventas Anuales')
plt.ylabel('Venta total')
plt.show()
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
total=[]
produ= []
aux=0
for i in range(1, len(df.keys())-2):
  produ.append(df.keys()[i])
  for n in range(0, len(df[df.keys()[i]])):
    aux+=df[df.keys()[i]][n]
  total.append(aux)

plt.pie(total,labels=produ)
plt.show()
```
